```python
import pandas as pd
import seaborn as sns
import numpy as np
dataset1 = pd.read_csv('/content/2019-oct.csv')
dataset2 = pd.read_csv('/content/2019-nov.csv')
dataset3 = pd.read_csv('/content/2019-dec.csv')
dataset1
```

```python
display(dataset1.shape,dataset2.shape,dataset3.shape)
```

```python
merged_dataset = pd.concat([dataset1, dataset2, dataset3])
merged_dataset.shape
```
